Genero una nueva variable y la llamo dfas, en la misma unicamente voy a cargar las columnas que me interesan. Algunas numericas y otras categoricas. El objetivo es utilizar unicamente las columnas utiles para las hipotesis planteadas.

```python
dfas = dfa[['Country', 'Year', 'Units', 'Category', 'Sales Size Sort', 'Unit Cost', 'Unit Price', 'SalesTax', 'PurchAgain']]
```

```python
dfas.info()
```

En este paso transformo algunas variables categoricas en numericas, para que cuando corra el modelo de regresion, basicamente no de error.

```python
dfas['Country'] = dfas['Country'].astype('category')
dfas['Country'] = dfas['Country'].cat.codes
dfas['Category'] = dfas['Category'].astype('category')
dfas['Category'] = dfas['Category'].cat.codes
dfas
```

 En esta instancia definimos una nueva variable que contenga variables numericas unicamente con las mismas columnas definidas, menos year que se utilizara como parametro clave en la otra variable necesaria para generar el modelo de regresion,

```python
df_asx = dfas.copy()
df_asx = df_asx.drop(columns= ['Year'])
df_asx
```

En este paso, normalizo los valores numericos para poder realizar calculos. Reemplazando o eliminando puntos o comas segun corresponda.

```python
df_asx['SalesTax'] = df_asx['SalesTax'].str.replace('.', '')
df_asx['SalesTax'] = df_asx['SalesTax'].str.replace(',', '.').astype(float)
```

**Procesamiento de datos**

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
dfas_y = dfas[['Year']]
dfas_y
```

```python
columnas_a_procesar = ['Unit Cost', 'Unit Price']

df_asx[columnas_a_procesar] = df_asx[columnas_a_procesar].apply(lambda x: x.str.replace(',', '.')).astype(float)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df_asx,dfas_y,test_size=0.3,random_state=2)
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
```

**Insights y recomendaciones**

¿ Bajo que circunstancia tendriamos un mejor rendimiento ?

Teniendo en cuenta las variables procesadas en x e y, llegamos a la conclusion que es probable que se generen mayor cantidad de ventas en un año con caracteristicas similares al 2013/2014

```python
print(y_test['Year'].mean())
```

```python
from sklearn.metrics import mean_absolute_error
print("MAE",mean_absolute_error(y_test,y_pred))
```

```python
from sklearn.metrics import mean_squared_error
print("MSE",mean_squared_error(y_test,y_pred))
```

```python
print("RMSE",np.log(np.sqrt(mean_squared_error(y_test,y_pred))))
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test,y_pred)
print(r2)
```
